```python
MAX_FEATURES = 200000 # no of words in the vocab
```

```python
vectorizer = TextVectorization(max_tokens = MAX_FEATURES,
                              output_sequence_length = 1000,
                              output_mode = 'int')
```

```python
# Force TensorFlow to Use CPU (if GPU is Not Needed)
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
```

```python
# teaching vectorizer our vocab
vectorizer.adapt(x.values)
```

```python
vectorizer.get_vocabulary()
```

```python
vectorized_text = vectorizer(x.values)
```

```python
len(x)
```

```python
# create tensorflow data pipeline
# MCSHBAP - map,cache,shuffle,batch,prefetch
# how to instantiate data pipelines?? two ways -> from_tensor_slices, list_dir

dataset = tf.data.Dataset.from_tensor_slices((vectorized_text,y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8) # helpts prevent bottlenecks
```

```python
batch_x, batch_y = dataset.as_numpy_iterator().next()
```

```python
# len(dataset)
int(len(dataset) * .7)
```

```python
train = dataset.take(int(len(dataset) * .7))
val = dataset.skip(int(len(dataset) * .7)).take(int(len(dataset) * .2))
test = dataset.skip(int(len(dataset) * .9)).take(int(len(dataset) * .1))
```

```python
len(train)
```

```python
train.as_numpy_iterator().next()
```

```python
# Create Sequential Model (deep learning model)
```

```python
from tensorflow.keras import Input,Sequential
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional, Dense, Embedding
```

```python
# model = Sequential()
# # create the embedding layer
# model.add(Embedding(MAX_FEATURES+1, 32))
# # bidirectional LSTM layer
# model.add(Bidirectional(LSTM(32, activation='tanh')))
# # Feature extractor fully connected layers
# model.add(Dense(128, activation='relu'))
# model.add(Dense(256, activation='relu'))
# model.add(Dense(128, activation='relu'))
# # final layer
# model.add(Dense(6, activation='sigmoid'))

model = Sequential([
    Input(shape=(1000,)),  # Each input is a sequence of 1000 tokens
    Embedding(MAX_FEATURES + 1, 32),
    (LSTM(32, activation='tanh')),
    Dense(128, activation='relu'),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(6, activation='sigmoid')
])
```

```python
Bidirectional??
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.summary()
```

```python
# Train the model
history = model.fit(train, epochs=10, validation_data=val)

# Save the trained model
model.save('my_model.h5')
print("Model saved successfully!")
```

```python
# saving model in keras format
model.save('my_model.keras')
print("Model saved successfully in Keras format!")

```

```python
# EVALUATE THE MODEL
```

```python
from tensorflow.keras.metrics import Precision, Recall, CategoricalAccuracy
```

```python
pre = Precision()
re = Recall()
acc = CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    # unpack the batch
    X_true, y_true = batch
    # make a prediction
    yhat = model.predict(X_true)

    # flatten the predictions
    y_true = y_true.flatten()
    yhat = yhat.flatten()
    
    pre.update_state(y_true, yhat)
    re.update_state(y_true, yhat)
    acc.update_state(y_true, yhat)
```

```python
print(f"""
Precision: {pre.result().numpy()},
Recall: {re.result().numpy()},
Accuracy: {acc.result().numpy()}
""")
```

```python
def score_comment(comment):
    vectorized_comment = vectorizer([comment])
    results = model.predict(vectorized_comment)

    text = ''
    for idx, col in enumerate(df.columns[2:]):
        text += '{}: {}\n'.format(col, results[0][idx] > 0.5)

    return text
```

```python
interface = gr.Interface(fn=score_comment,
                         inputs=gr.Textbox(lines=2, placeholder='Comment to score'),
                         outputs='text')
```
